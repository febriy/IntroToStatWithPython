```python
import numpy as np
import seaborn as sns
```

```python
sample_dataset = np.random.normal(50,5, 1000)
```

```python
ax = sns.distplot(sample_dataset)
```

```python
ax = sns.distplot(sample_dataset, rug=True, hist=False)
```

- Rugplot (rug): One-dimensional plot of the variables, plotted on the axis. Represents density.
- [Kernel Density Estimate](https://mathisonian.github.io/kde/) (kde): Algorithm used to create smooth line from underlying dataset. Useful if we want to see the 'shape' of the distribution.
