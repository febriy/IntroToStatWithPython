```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
eve = pd.read_csv('001_Everion_data.csv')
eve['date'] = pd.to_datetime(eve['Timestamp'],unit='s')
eve.head()
```

```python
eve.plot(x='date', y='Value', figsize=(15,5))
```

```python
om = pd.read_csv('001_OMRION_data.csv')
om['date'] = pd.to_datetime(om['Timestamp'],unit='s')
om.head()
```

```python
om.plot(x='date', y='Value', figsize=(15,5))
```

```python
plt.figure(figsize=(15,5))
plt.plot( 'date', 'Value', data=eve, marker='', color='olive', linewidth=2, label="everion")
plt.plot( 'date', 'Value', data=om, marker='', color='blue', linewidth=2, linestyle='dashed', label="omrion")
plt.legend()
```
